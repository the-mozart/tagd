```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
# Включение русского языка
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)
rc('text.latex',unicode=True)
rc('text.latex',preamble='\\usepackage[utf8]{inputenc}')
rc('text.latex',preamble='\\usepackage[russian]{babel}')
```

```python
%matplotlib notebook

w = 3
Y, X = np.mgrid[-w:w:300j, -w:w:300j]
U = -Y / (X**2 + Y**2)
V = X / (X**2 + Y**2)
speed = np.sqrt(U*U + V*V)


fig, ax = plt.subplots(1,1,figsize=(5,5))

# gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

# ax1 = fig.add_subplot(gs[0, 1])
strm = ax.streamplot(X, Y, U, V, color=U, linewidth=1)

ax.grid(True)

ax.set_xlabel("Re")
ax.set_ylabel("Im")


plt.savefig("rot_Gamma_g_0.pdf")
```

```python
%matplotlib notebook

w = 3
Y, X = np.mgrid[-w:w:300j, -w:w:300j]

COST = X / (X**2 + Y**2)
SINT = Y / (X**2 + Y**2)

COS2T = COST**2-SINT**2
SIN2T = 2 * SINT * COST
r = np.sqrt(X**2+Y**2)


U = -1/r**2 * COS2T
V = -1/r**2 * SIN2T

speed = np.sqrt(U*U + V*V)


fig, ax = plt.subplots(1,1,figsize=(5,5))

# gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

# ax1 = fig.add_subplot(gs[0, 1])
strm = ax.streamplot(X, Y, U, V, color=U, linewidth=1)

ax.grid(True)

ax.set_xlabel("Re")
ax.set_ylabel("Im")


plt.savefig("dipol.pdf")
```

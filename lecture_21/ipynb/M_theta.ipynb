```python
import matplotlib.pyplot as plt
import numpy as np

# Включение русского языка
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('font',**{'size':'18'})
rc('text', usetex=True)
rc('text.latex',unicode=True)
rc('text.latex',preamble='\\usepackage[utf8]{inputenc}')
rc('text.latex',preamble='\\usepackage[russian]{babel}')

```

```python
gamma = 1.4

def theta(M):
    return M/((2/(gamma+1)*(1+(gamma-1)/2*M*M))**((gamma+1)/2/(gamma-1)))

Mi = np.linspace(0.01,4)

fig, ax = plt.subplots(1,1,figsize=(6,5))

ax.set_xlabel("$M$")
ax.set_ylabel("$\\Theta$")

ax.set_xlim(0, 4)
# ax.set_ylim(0,1)

ax.plot(Mi,theta(Mi))

ax.plot(0.2, theta(0.2),'ro')
ax.plot(0.8, theta(0.8),'ro')

ax.plot(1, theta(1),'go')
ax.plot(3.2, theta(3.2),'go')

ax.grid(True)

plt.tight_layout()

plt.savefig("../img/M_theta.pdf")
# ax.legend(loc='best')   
```

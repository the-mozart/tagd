## Обтекание полой пористой сферы вязкой несжимаемой жидкостью

Рассмотрим задачу стационарного обтекания полого пористого шара потоком вязкой несжимаемой жидкости с внешним радиусом $b$ и радиусом полости $a$. Скорость потока на бесконечности будем полагать равной $v_{\infty}$ (рис. 1), $\mu$, $\rho$ -- динамическая вязкость и плотность жидкости.

Введем сферическую систему координат $(r,\theta,\varepsilon)$. Тогда можно считать, что искомые параметры течения радиальная составляющая скорости $v_r$, нормальная составляющая скорости $v_{\theta}$ и давление в жидкости $p$ являются функциями только параметров $r$ и $\theta$. Также можно пренебречь азимутальной составляющей скорости $v_{\varepsilon}$.

Рассмотрим сечение шара плоскостью, проходящей через точку $O$. Т.к. течение осесимметрично, то можно считать, что это плоскость $Oxz$.
Преобразования координат в этой плоскости соответстуют полярной системе кооридинат
$$
\left\{
\begin{array}{l}
x = r \sin \theta, \\
z = r \cos \theta,\\
\end{array}\right.,\quad
\left\{
\begin{array}{l}
r = \sqrt{x^2+z^2}, \\
\theta = \theta(|x|, z).\\
\end{array}\right. ,
$$
$$
0 \leq r <\infty,\quad
0 \leq \theta \leq \pi,\quad
x \geq 0.
$$

Преобразование векторов из введеной системы координат в ортогональную имеет вид
$$
\begin{array}{l}
v_x = v_r \sin\theta + v_{\theta} \cos\theta,\\
v_z = v_r \cos\theta - v_{\theta} \sin\theta.\\
\end{array}
$$

Предположим, что свободное течение описывается в рамках приближения Стокса, которое справедливо для течений с числами Рейнольдса ${\rm Re} << 1$ $({\rm Re} = 2b v_{\infty} \rho / \mu)$, а фильтрационное течение в стенке шара подчиняется закону Дарси. 

Уравнения Стокса свободного течения имеют вид
$$
\pt{v_r}{r}+\frac{1}{r}\pt{v_{\theta}}{\theta}+\frac{2}{r}v_r+\frac{\ctg \theta}{r}v_{\theta} = 0,
$$
$$
\mu\left(\ppt{v_r}{r} + \frac{1}{r^2}\ppt{v_r}{\theta} + \frac{2}{r}\pt{v_r}{r} + \frac{\ctg \theta}{r^2}\pt{v_r}{\theta} - \frac{2v_r}{r^2} - \frac{2}{r^2}\pt{v_{\theta}}{\theta} -  \frac{2 \ctg \theta}{r^2}v_{\theta}  \right)  =  \pt{p}{r},
$$
$$
\mu\left(\ppt{v_{\theta}}{r} + \frac{1}{r^2}\ppt{v_{\theta}}{\theta} +  \frac{2}{r}\pt{v_{\theta}}{r} + \frac{\ctg \theta}{r^2}\pt{v_{\theta}}{\theta} + \frac{2}{r^2}\pt{v_r}{\theta} - \frac{v_{\theta}}{r^2 \sin^2 \theta} \right) = \frac{1}{r}\pt{p}{\theta},
$$
$$
0< r < a, \quad r>b, \quad 0\leq\theta\leq\pi.
$$

Уравнения фильтрационного течения Дарси имеют вид
$$
\pt{V_r}{r}+\frac{1}{r}\pt{V_{\theta}}{\theta}+\frac{2}{r}V_r+\frac{\ctg \theta}{r}V_{\theta} =  0,
$$
$$
V_r = -\frac{k}{\mu}\pt{P}{r},
$$
$$
V_{\theta} = -\frac{k}{\mu}\frac{1}{r}\pt{P}{\theta}
$$
$$
	a\leq r\leq b, \quad 0\leq\theta\leq\pi,
$$
где $V_{r}$ -- радиальная составляющая скорости фильтрации, $V_{\theta}$ -- нормальная составляющая скорости фильтрации, $P$ -- давление. Далее параметры течения в полости шара обозначаются $v_r^{i}$, $v_{\theta}^{i}$, $p^{i}$, вне шара -- $v_r^{e}$, $v_{\theta}^{e}$, $p^{e}$.

Зададим условия на границах между областями свободного и фильтрационного течений, которые представляют собой условия прилипания вязкой жидкости на границе с твердым телом и условие непрерывности давления:

-- при $r=a$
$$
v_r^i=V_r,\quad \pt{v_{\theta}^i}{r}=\frac{\alpha}{\sqrt{k}}(v_{\theta}^i - V_{\theta}),\quad p^i=P;
$$

-- при $r=b$
$$
v_r^e=V_r,\quad v_{\theta}^e=0, \quad p^e=P,
$$

-- а также условие на бесконечности
$$
v_r^e \to   -v_{\infty} \cos\theta,\quad v_{\theta}^e \to v_{\infty} \sin\theta.
$$

Рассмотрим задачу Стокса, общее решение которой имеет вид
$$
v_r  =  \left(\frac{C_1}{r^3}+\frac{C_2}{r}+C_3+C_4r^2\right)\cos\theta,
$$
$$
v_{\theta} =  \left(\frac{C_1}{2r^3}-\frac{C_2}{2r}-C_3-2C_4r^2\right)\sin\theta,
$$
$$
p  = \mu\left(\frac{C_2}{r^2}+10C_4r\right)\cos\theta.
$$

Проверим, что это действительно решение указанных уравнений.

````python
def lambify_vxvz_from_vrvtheta(vr, vtheta, unknowns_from_params, params):

    vx, vz = vector_to_xz(vr, vtheta)
    
    return sp.lambdify((x, z), vx.subs(unknowns_from_params).subs(params).simplify().subs(rtheta_to_xz).simplify(), 'numpy'), \
           sp.lambdify((x, z), vz.subs(unknowns_from_params).subs(params).simplify().subs(rtheta_to_xz).simplify(), 'numpy')

def velocity_external(unknowns_from_params, params):
    return lambify_vxvz_from_vrvtheta(vr_e, vtheta_e, unknowns_from_params, params)

def velocity_internal(unknowns_from_params, params):
    return lambify_vxvz_from_vrvtheta(vr_i, vtheta_i, unknowns_from_params, params)

def velocity_filtration(unknowns_from_params, params):
    return lambify_vxvz_from_vrvtheta(Vr, Vtheta, unknowns_from_params, params)
````

````python
vx_e, vz_e = velocity_external(sol, params) # внешнее поле скорости
vx_i, vz_i = velocity_internal(sol, params) # внутреннее поле скорости
vx_f, vz_f = velocity_filtration(sol, params) # фильтрационное поле скорости

def velocity(x, z):
    R = np.sqrt(z**2+x**2)

    vx = np.zeros_like(x)
    vz = np.zeros_like(z)

    inside     = (R <= params[a])
    outside    = (R >= params[b])
    filtration = (R > params[a]) & (R < params[b])

    print(inside.sum(), filtration.sum(), outside.sum())
    
    vx[inside] =  vx_i(x[inside], z[inside])
    vz[inside] =  vz_i(x[inside], z[inside])

    vx[outside] =  vx_e(x[outside], z[outside])
    vz[outside] =  vz_e(x[outside], z[outside])

    vx[filtration] =  vx_f(x[filtration], z[filtration])
    vz[filtration] =  vz_f(x[filtration], z[filtration])

    return vx, vz
````

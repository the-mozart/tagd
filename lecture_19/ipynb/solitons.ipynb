```python
import matplotlib.pyplot as plt
import numpy as np

# Включение русского языка
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)
rc('text.latex',unicode=True)
rc('text.latex',preamble='\\usepackage[utf8]{inputenc}')
rc('text.latex',preamble='\\usepackage[russian]{babel}')
```

```python
# %matplotlib notebook

def f(xi, x0, delta):
    return np.exp(-(xi-x0)**2/2/delta**2)

x = np.linspace(-2,2, num = 200)

x01 = -0.75
sigma1 =  1/5

x02 = 0.75
sigma2 = 1/10

c0 = 1/5


for t in np.arange(1,8):
    f1 = f(x-c0*t, x01, sigma1)
    f2 = f(x+c0*t, x02, sigma2)

    fig, ax = plt.subplots(1,1,figsize=(10,5))

    ax.plot(x,f1,'r',label = "$f_1$")
    ax.plot(x,f2,'b',label = "$f_2$")

    ax.plot(x,f1+f2,'g--',label = "$\\varphi$")
    
#     ax.arrow(, f(-0.75+0.1 - t ,-0.75, 1/5), 0.1, 0,head_width=0.05, head_length=0.05, fc='k', ec='k') 
#     ax.arrow(x01+sigma1/2 - t*c0, f(x01+sigma1/2 - t*c0, x01, sigma1), 0.2,0,head_width=0.05, head_length=0.05, fc='k', ec='k')
    
    xa1 = x01 + sigma1/2 + c0*t
    ya1 = f(x01 + sigma1/2, x01, sigma1)

    xa2 = x02 - sigma2 - c0*t
    ya2 = f(x02 - sigma2, x02, sigma2)
    
    ax.arrow(xa1,ya1, 0.2,0,head_width=0.05, head_length=0.05, fc='k', ec='k')
    ax.arrow(xa2,ya2, -0.2,0,head_width=0.05, head_length=0.05, fc='k', ec='k')
#     ax.plot(xa1 , ya1,'o')
    
    ax.grid(True)
    ax.legend(loc='best')

    ax.set_xlabel("x")
    ax.set_ylabel("$\\varphi$")

    ax.set_xlim(-1.2,1.2)
    ax.set_ylim(0,2)
      
   
        
    plt.savefig("img/signal_%02d.pdf" % t)
    plt.close()
```

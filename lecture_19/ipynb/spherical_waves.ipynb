```python
import matplotlib.pyplot as plt
import numpy as np

# Включение русского языка
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)
# rc('text.latex',unicode=True)
rc('text.latex',preamble='\\usepackage[utf8]{inputenc}')
rc('text.latex',preamble='\\usepackage[russian]{babel}')
```

```python
def f(ri, r0, delta):
    return np.exp(-(ri-r0)**2/2/delta**2)

r = np.linspace(0.01,3, num = 200)

r01 = 0.5
sigma1 =  1/5

r02 = 2.5
sigma2 = 1/10

c0 = 0.1


for t in np.arange(5.5,15,1):
    f1 = f(r-c0*t, r01, sigma1)/r
    f2 = f(r+c0*t, r02, sigma2)/r

    fig, ax = plt.subplots(1,1,figsize=(8,4))

    ax.plot(r,f1,'r',label = "$f_1/r$")
    ax.plot(r,f2,'b',label = "$f_2/r$")

    ax.plot(r,f1+f2,'g--',label = "$\\varphi$")
    
    
    ra1 = r01 + sigma1/2 + c0*t
    ya1 = f(r01 + sigma1/2, r01, sigma1)/ra1
    
    ra2 = r02 - sigma2 - c0*t
    ya2 = f(r02 - sigma2, r02, sigma2)/ra2
    
    ax.arrow(ra1,ya1, 0.2,0,head_width=0.03, head_length=0.05, fc='k', ec='k')
    ax.arrow(ra2,ya2, -0.2,0,head_width=0.03, head_length=0.05, fc='k', ec='k')

    ax.grid(True)
    ax.legend(loc='best')

    ax.set_xlabel("$r$")
    ax.set_ylabel("$\\varphi$")

    ax.set_xlim(0.01,2.6)
    ax.set_ylim(0,1.5)
      
   
        
    plt.savefig("img/spherical_signal_%02d.pdf" % t)
#     plt.close()
```
